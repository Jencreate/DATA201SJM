```r
library(readxl)
install.packages("openxlsx")
library(openxlsx)
electricity_cost <- 
read.xlsx("https://www.mbie.govt.nz/assets/Data-Files/Energy/nz-energy-quarterly-and-energy-in-nz/qsdep-report-15-nov-2020.xlsx",
          sheet = "RawData")
electricity_cost # Open excel file from web
```

```r
electricity_cost_2020 <- electricity_cost %>%
subset(select = c("Retail.(c/kWh)","43876", "43966", "44058", "44150"))
electricity_cost_2020 # Select necessary data
```

```r
as.Date(c(43876, 43966, 44058, 44150), origin="1899-12-30")
electricity_cost_2020 <- electricity_cost_2020 %>% format(as.Date(c(43876, 43966, 44058, 44150), origin="1899-12-30"))
electricity_cost_2020
```
