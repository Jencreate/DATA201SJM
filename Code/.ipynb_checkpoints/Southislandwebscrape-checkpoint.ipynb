```r
library(tidyverse)
#importing libraries for use in this part, only tidyverse library required
```

```r
url <- "https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_the_South_Island" #base url to scrape town names from
page_title = read_html(url) #reading in the url using read_html from the rvest package in tidyverse 
output <- page_title %>% html_nodes(".div-col")%>% html_text()%>% str_split("\n") #obtained multiple lists of town names, grouped alphabetically
#used html_nodes and html_text from rvest to extract the text from the div-col attributes in the webpage, then used str_split to remove the new line and seperate by a comma
```

```r
list_towns <- list() #initilising an empty list to append to
#print(list_towns) #making sure it had initilised correctly
#print(output) #viewing output to figure out how to index correctly
for(i in 1:19) { #looping through 19 times, as this is the number of unique alphabetical letters in the town names
    list_towns <- list_towns %>% append(output[[i]][c(-1)]) #indexing into the alphabetical letter and extracting each town
}
```
